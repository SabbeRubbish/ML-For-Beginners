# Diabetes lesson

````python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
````

## Load data from SciKit
Dataset is diabetes illness progression (y) based on 442 patients with the following variables:
```
AGE SEX BMI BP S1 S2 S3 S4 S5 S6 Y
```
In this example, we'll use the ``BMI`` variable as an example

````python
X, y = datasets.load_diabetes(return_X_y=True)
# Show histogram of BMI
variable_to_use = 6 # set from 0 to 9
plt.hist(X[:, variable_to_use], bins=100)
plt.gca().set(ylabel="Frequencey of ages (X)", xlabel="BMI")
````

The y array shows the diabetes progression based on the parameters.

We can't really plot that in any significant way.

````python
X = X[:, np.newaxis, variable_to_use]
````

## Data split
We're splitting the dataset (X and y) into 2 pools, one to train, one to test.

Split is done at the 1/3 mark, resulting in sets of 296 (2/3) and 146 (1/3) mark.

````python
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
````

## Training
Now we're going to train our data using linear regression.


<img src="images/regression.png" style="text-align:center" />

````python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
````

Our model has been trained now with the test (296 records) data.

Now we'll predict how the test data will behave (remaining 146 records).

````python
y_pred = model.predict(X_test)
````

I've added the training data as well (pink) for more insight.

````python
plt.scatter(X_train, y_train,  color='pink')
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
````

# Conclusion
All variables
```
AGE SEX BMI BP S1 S2 S3 S4 S5 S6 Y
```
have a positive influence in the disease progression (y) **except** variable S3 (high-density lipoproteins)
